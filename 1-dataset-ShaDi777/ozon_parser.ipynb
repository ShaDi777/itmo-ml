### Преобразуем столбец Разрешение

```python
df['Разрешение'].unique()
```

```python
known_resolutions = {
    '1080P': '1920x1080',
}

def replace_known_resolutions(value):
    for key, resolution in known_resolutions.items():
        if key in value:
            return resolution
    return value

df['Разрешение'] = df['Разрешение'].fillna('')
df['Разрешение'] = df['Разрешение'].apply(replace_known_resolutions)

df[['Ширина, пиксели', 'Высота, пиксели']] = df['Разрешение'].str.extract(r'(\d{3,5})x(\d{3,5})')

df['Ширина, пиксели'] = pd.to_numeric(df['Ширина, пиксели'], errors='coerce')
df['Высота, пиксели'] = pd.to_numeric(df['Высота, пиксели'], errors='coerce')

df = df.drop(columns=['Разрешение'])
df.to_csv('step1_resolution.tsv', sep='\t', index=False)
df
```

### Далее будем разбивать столбец Назначение монитора

```python
df = pd.read_csv('step1_resolution.tsv', sep='\t')
df['Назначение монитора'].unique()
```

```python
df['Назначение монитора'] = df['Назначение монитора'].str.split(',')
df['NaN_Назначение'] = df['Назначение монитора'].isna()

df_exploded = df.explode('Назначение монитора')
df_encoded = pd.get_dummies(df_exploded, columns=['Назначение монитора'])

aggregation_dict = {}
for col in df_encoded.columns:
    if 'Назначение монитора' in col:
        aggregation_dict[col] = 'max'
    else:
        aggregation_dict[col] = 'first'

df_final = df_encoded.groupby(level=0).agg(aggregation_dict)

for col in df_final.columns:
    if 'Назначение монитора' in col:
        df_final[col] = df_final[col].where(~df['NaN_Назначение'], None)
df_final = df_final.drop(columns='NaN_Назначение')

df_final.to_csv('step2_usage.tsv', sep='\t', index=False)

df_final
```

```python
def parse_contrast(value):
    if pd.isna(value):
        return None

    value = value.lower().replace(' ', '').replace('m', '000000').replace('м', '000000')

    value = re.sub(r'[^\d\.:]', '', value)

    parts = re.split(r':|：|/|к', value)
    numbers = [int(p) for p in parts if p.isdigit()]

    return max(numbers) if numbers else None

df['Контрастность'] = df['Контрастность'].apply(parse_contrast)
df['Динамическая контрастность'] = df['Динамическая контрастность'].apply(parse_contrast)

df.to_csv('step5_contrast.tsv', sep='\t', index=False)

df
```

```python
def parse_viewing_angles(value):
    if pd.isna(value):
        return None, None

    value = value.replace('°', '')
    horizontal, vertical = value.split('/')

    return int(horizontal), int(vertical)

df[['Угол обзора горизонтальный', 'Угол обзора вертикальный']] = df['Углы обзора (Г/В)'].apply(parse_viewing_angles).apply(pd.Series)

df.drop('Углы обзора (Г/В)', axis=1, inplace=True)

df.to_csv('step6_sight_angle.tsv', sep='\t', index=False)

df
```

```python
hdr_tier = {
    'HDR10': 1,
    'HLG': 2,
    'DisplayHDR 400': 3,
    'DisplayHDR True Black 400': 3,
    'HDR10+': 4,
    'DisplayHDR 600': 5,
    'DisplayHDR 1000': 6,
    'DisplayHDR 1400': 7,
    'Dolby Vision': 8,
    'DisplayHDR 1600': 9,
    'Dolby Vision IQ': 10,
}

def get_max_hdr_tier(hdr_values):
    if pd.isna(hdr_values):
        return None
    hdr_items = hdr_values.split(',')
    hdr_scores = [hdr_tier[item] for item in hdr_items if item in hdr_tier]
    return max(hdr_scores) if hdr_scores else None

df['Cтандарты HDR'] = df['Cтандарты HDR'].apply(get_max_hdr_tier)
df.to_csv('step9_hdr.tsv', sep='\t', index=False)

df
```

```python
def clean_and_split_dimensions(dimensions):
    if pd.isna(dimensions):
        return [None, None, None]

    matches = re.findall(r'\d+\.?\d*', dimensions)

    numbers = [float(num) for num in matches[:3]]

    while len(numbers) < 3:
        numbers.append(None)
    
    return numbers

df[['Ширина, мм', 'Высота, мм', 'Глубина, мм']] = pd.DataFrame(df['Размеры, мм'].apply(clean_and_split_dimensions).to_list(), index=df.index)
df.drop('Размеры, мм', axis=1, inplace=True)

df.to_csv('step13_dimensions.tsv', sep='\t', index=False)

df
```

### Преобразуем страну

```python
df = pd.read_csv('step14_brand.tsv', sep='\t')
```

```python
df['Страна-изготовитель'].unique()
```

```python
df['Страна-изготовитель'] = df['Страна-изготовитель'].replace('Не указана', None)

def split_countries(value):
    if pd.isna(value):
        return []
    return [country.strip() for country in value.split(',')]

df['Страна-изготовитель'] = df['Страна-изготовитель'].replace(
    {'Китай (Тайвань)': 'Китай', 'Китай (Гонконг)': 'Китай'}
)

df['Страна-изготовитель_Все'] = df['Страна-изготовитель'].apply(split_countries)

all_countries = df['Страна-изготовитель_Все'].explode()
country_freq = all_countries.value_counts()
country_mapping = {country: i+1 for i, country in enumerate(country_freq.index[::-1])}

def map_country_list(country_list):
    if not country_list:
        return 0
    return max([country_mapping.get(country, 0) for country in country_list])

df['Страна-изготовитель'] = df['Страна-изготовитель_Все'].apply(map_country_list)

df.drop(columns=['Страна-изготовитель_Все'], inplace=True)

df.to_csv('step15_country.tsv', sep='\t', index=False)

df
```

### Преобразуем Гарантийный срок

```python
df = pd.read_csv('step15_country.tsv', sep='\t')
```

```python
df['Гарантийный срок'].unique()
```

```python
def convert_to_months(value):
    if pd.isna(value):
        return None

    value = str(value).lower()

    if 'лет' in value or 'год' in value or 'year' in value or '年' in value:
        years = re.findall(r'\d+', value)
        if years:
            return int(years[0]) * 12
        if 'один' in value or 'первый' in value or 'oдnh' in value:
            return 12
        if 'три' in value:
            return 36
    elif 'месяц' in value or 'мес' in value or '个月' in value:
        months = re.findall(r'\d+', value)
        if months:
            return int(months[0])
    elif 'дни' in value or 'дней' in value or 'день' in value or 'days' in value or 'day' in value or '天' in value:
        days = re.findall(r'\d+', value)
        if days:
            return int(days[0]) // 30
    elif re.match(r'\d+(\.\d+)?', value):
        return float(value) if '.' in value else int(value)

    return None

df['Гарантийный срок, месяцы'] = df['Гарантийный срок'].apply(convert_to_months)

df.drop(columns=['Гарантийный срок'], inplace=True)

df.to_csv('step16_warranty.tsv', sep='\t', index=False)

df
```
